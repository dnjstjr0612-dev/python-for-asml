### Consider the 7-th order Taylor polynomial approximation of the function f(x) = sinx about x_0 = 0

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial
```

```python
# Taylor eq. definition
def taylor_sin(x, num_terms):
    """
    Args:
        x (np.ndarray): array
        num_terms (int): nth
        
    Returns:
        np.ndarray 
    """
    y = np.zeros_like(x)

    for n in range(num_terms):
        term = (-1)**n * x**(2*n + 1) / factorial(2*n + 1)
        y += term
    return y

x = np.arange(0, 2*np.pi, 0.01) # x 값 생성
y_sin = np.sin(x) # sin(x) 함수 계산과 그리기
plt.plot(x, y_sin, label = "sin(x), P_7(x)")

z = taylor_sin(x, 4) # plotting
plt.plot(x, z, label="P_7(x)")

plt.xticks(np.arange(0, 2*np.pi + 0.1, np.pi/2),
           ['0', 'π/2', 'π', '3π/2', '2π'])
plt.xlabel('x')
plt.ylabel('function')
plt.title('Plot of sin(x) and its Taylor Polynomial')
plt.legend()
plt.grid(True)
plt.show()
```
